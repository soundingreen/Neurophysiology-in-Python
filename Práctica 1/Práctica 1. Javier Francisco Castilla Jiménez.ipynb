## Funciones de importación de datos


```python
def abrir(archivo):
    with open(archivo, mode='r') as datos: # abrimos el archivo
        contenido= datos.readlines() #contenido aqui es una lista de strings 
    return contenido
def procesado(contenido):
    ensayos=[] # declaro una lista vacía para guardar arreglos donde cada arreglo contendra los registros correspondientes a un ensayo  ensayos 
    for string in contenido:# hay 63 str en contenido1 tres de esos str son \n  y los demás son los datos de 60 ensayos
        if string!='\n': # solo 60 str de contenido cumplen esta condición y esos son ensayos
            lista_n=[] # entonces declararé en total 60 listas vacías una en cada ciclo
            string=string.replace(','," ") # reemplazo las comas por espacios para poder usar split después
            string=string.split() #split hace cada número dentro del str de ensayos un str individual, y pone estos str dentro de una lista # ahora la variable string es una lista donde cada elemento es un str de un número flotante
            for numero in string: # itero cada elemento de la lista string
                numero=float(numero) # cada str de un número lo convierto a float 
                lista_n.append(numero)# cada float agrego con append a lista_n
            ensayo_n=np.array(lista_n)# al final de el segundo ciclo for agrego la lista_n correspondiente a un array que llamo ensayo_n
            ensayos.append(ensayo_n)# luego agrego mi arreglo ensayo_n a mi lista de arrays "ensayos"
    return ensayos

def importar(archivo): # función para importar datos
    content=abrir(archivo)
    print(content)
    Datos=procesado(content)
    return Datos
```

# Mis funciones
### para no repetir todo el código anterior en cada caso e inciso generalizaré lo que hice como funciones

```python
#Start es en donde empieza mi rango # en segundos
#Stop en donde termina rango #segundos
#paso #en segundos
#ventana # en segundos

def bordesup(start,stop,paso,ventana): # función para obtener los bordes superiores de mi ventana dadas las especificaciones de cada caso
    rangosup=float(np.abs((start+ventana)-stop)) 
    npuntos=int(rangosup/paso) # numero de puntos necesarios para tener los puntos adecuados
    x=np.linspace(start+ventana,stop,npuntos,endpoint=True)# calculo bordes superiores dada mi ventana 
    return x
```

```python
# aquí estamos trabajando unicamente con la clase 1 de equivale a n1[0:15] es decir los primeros quince ensayos
def frecuencia(datos,sup,anchov):
    frecuencias=[]
    for ensayo in (datos): #itero un array de ensayos a la vez
        fre_ensayo_n=[]# voy a generar tantas de estas lista como ensayos haya
        for borde in sup:# itero los elementos de mi array de bordes, es decir itero los bordes superiores
            frec=int(np.sum(np.logical_and((ensayo>=(borde-anchov)),(ensayo<borde)))) # la cantidad de elementos en un ensayo que están dentro de la ventana definida por esos bordes 
            fre_ensayo_n.append(frec) # está lista tiene los frecuencias de cada ventana
        frecssay=np.array(fre_ensayo_n) # este es un array que tiene las frecuencias de cada ventana para un ensayo_n
        frecuencias.append(frecssay) # frecuencias es una lista de arrays donde cada array tiene las frecuencias por ventana de cada ensayo
    return frecuencias
```

```python
def frate(frecuencias,anchov):
    fdv=0
    for array in frecuencias: #
        fdv=fdv+array # sumo todos los arrays de frecuencias para obtener un único array con las suma de todas las frecuencias donde cada elemnto corresponde a una vnetana distinta
    tasaD=((fdv/anchov)/len(frecuencias))# tasa D es un array que contiene las tasas de disparo por ventana de todos los  ensayos de esta clase 
    return tasaD
```

```python
##por último una función para resolver los ejercicios casi en una sola linea
def tasa(datos,sup,anchov):
    frecuencias=frecuencia(datos,sup,anchov)
    tasadisparo=frate(frecuencias,anchov)
    return tasadisparo
```
