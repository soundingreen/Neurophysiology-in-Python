```python
import numpy as np 
import matplotlib.pyplot as plt
import NeuralFunctJC as nrn
plt.style.use('Solarize_Light2')
```

```python
llaves=['1','3','5']
#Avión, Barco, Bicicleta, Tren.
#cargamos los datos
AG1=nrn.importar('ArchivoG1.csv')
AG3=nrn.importar('ArchivoG3.csv')
AG5=nrn.importar('ArchivoG5.csv')
Arch={'1':np.array(AG1),'3':np.array(AG3),'5':np.array(AG5)}
```

```python
#determino mis bins para cada set de datos
nbins=20
bines=dict()
for llave in llaves: # un ciclo para cada llave en llaves
    m,M=(np.min(np.array(Arch[llave]).flatten()),np.max(np.array(Arch[llave]).flatten()))#busco min y max de un set de datos #mínimo y máximo
    bins=np.linspace(round(m)-1,round(M)+1,nbins+1)#cálculo de bins
    bines.update({llave:bins})
############################
```

```python
llaves2=['avion','barco','bicicleta','tren'] #llaves de estímulo
ProbArc=dict() # diccionario de diccionarios de probabilidades por tipo de estímulo por neurona
for llave in llaves: #un ciclo para cada llave en llaves
    AG=dict()# diccionario vacío para probabilidades condicionadas y prob total
    for i in range(len(llaves2)):
        counts, bins=np.histogram(Arch[llave][i],bins=bines[llave]) #count guarda un array con la frecuencia y bins un array con los bordes de bins
        counts=counts/counts.sum()#cálculo de distribución de probabilidad #convertimos la frecuencia en probabilidad (frecuencia relativa)
        AG.update({llaves2[i]:[bins,counts]})
    #calculamos bines y probabilidades de la ddistribución total:
    total=np.concatenate(tuple(Arch[llave])) # todas las tasas de disparo sin estar condicionadas a un tipo de estímulo
    counts, bins=np.histogram(total,bins=bines[llave]) #count guarda un array con la frecuencia y bins un array con los bordes de bins
    counts=counts/counts.sum()#cálculo de distribución de probabilidad de la tasa de disparo #convertimos la frecuencia en probabilidad
    AG.update({'dtotal':[bins,counts]})#
    ProbArc.update({llave:AG})# diccionario de diccionarios 
```

```python
def I_shanon4(data,keys,ps,keytotal='dtotal'):
    """data es los datos correspondientes a un archivo, llaves=son las llaves de los estímulos
    ps=probabilidad de aparición del estímulo
    ptasa=probabilidad de la tasa
    pcondicionada= probabilidad de la tasa condicionada al estímulo"""
    ptasa=data[keytotal][1]
    Info_stim=[]
    for llave in keys:
        pcondicionada=data[llave][1]
        Is=np.nansum(pcondicionada*np.log2(pcondicionada/ptasa))
        Info_stim.append(Is)
    Ish=ps*np.sum(Info_stim)
    return Ish
```

```python
names={'1':'AG1','3':'AG3','5':'AG5'}
shanon=[]
for llave in llaves:
    Ishanon=I_shanon4(ProbArc[llave],llaves2,ps=1/4)
    shanon.append(Ishanon)
    print('la información mutua de '+names[llave]+' es:' ,Ishanon,'bits')
```
