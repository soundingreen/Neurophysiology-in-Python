```python
import numpy as np 
import matplotlib.pyplot as plt
import NeuralFunctJC as nrn
plt.style.use('Solarize_Light2')
%matplotlib inline
```

```python
########## cargamos datos
neuronas=['G01','G05','G1','G2','GS3']
n=[nrn.importar('NeuronaSimulada'+neuron+'.csv') for neuron in neuronas]
DatosG={neuronas[i]:n[i]for i in range(len(n))} # diccionario por neurona de lista de listas con los np.ndarrays de lod tiempos de ocurrencia de las espigas
```

```python
def ROC(sample1,sample2,bines=100):
    dist1,dist2=nrn.getDistributions(sample1,sample2,nbins=bines)
    auroc=nrn.getAUROC(dist1,dist2) 
    return auroc
```

```python
for neuron in neuronas:
    print('el AUROC de '+neuron+' es:',ROC(DatosG[neuron][0],DatosG[neuron][1]))
```

```python
def permutacion(data,nperm=1000,bines=100):
    forma=data.shape
    ndat=forma[0]*forma[1]
    pr_data=np.copy(data)
    pr_ROC=np.empty(nperm,dtype=float)# matriz vacía para los ROCS permutados
    indices=[np.random.permutation(ndat) for i in range(nperm)]#lista de los indices permutados para las mil permutaciones
    for b in range(nperm):# un ciclo por permutacion
        plano=data.flatten()[indices[b]]#datos aplanados y permutados de acuerdo a los indices correspondientes
        pr_data=plano.reshape(forma)# datos permutados y con la estructura/shape original
        pr_ROC[b]=ROC(pr_data[0],pr_data[1],bines)
    return pr_ROC
```

```python
def remuestreo(data,nc=2,nxc=200,nboots=1000,bines=100):
    bs_data=np.copy(data)
    bs_ROC=np.empty(nboots,dtype=float)
    indices=np.random.choice(nxc,size=(nboots,nc,nxc),replace=True)
    for b in range(nboots):
        for i in range(nc):
            bs_data[i]=data[i][indices[b][i]]
        bs_ROC[b]=ROC(bs_data[0],bs_data[1],bines)
    return bs_ROC
```

```python
for neuron in neuronas:
    A_real=ROC(DatosG[neuron][0],DatosG[neuron][0])
    A_permutada=permutacion(np.array(DatosG[neuron]))
    p=np.sum(A_permutada[A_permutada>A_real])/len(A_permutada)
    fig = plt.figure(figsize=(8,5))
    plt.hist(A_permutada,label='p='+str(p))
    plt.xlabel("bin",fontsize=14)
    plt.ylabel("frecuencia",fontsize=14)
    plt.title("Histograma AUROCs permutada "+'\n set '+neuron)
    plt.legend(fontsize="medium",shadow=True,fancybox=True)
    plt.show() 
```

```python
for neuron in neuronas:
    A_remuestreada=remuestreo(np.array(DatosG[neuron]))
    sigma=np.std(A_remuestreada)
    plt.hist(A_remuestreada,label='error=±{:.3f}'.format(2*sigma))
    plt.xlabel("bin",fontsize=14)
    plt.ylabel("frecuencia",fontsize=14)
    plt.title("Histograma AUROCs remuestreo "+'\n set '+neuron)
    plt.legend(fontsize="medium",shadow=True,fancybox=True)
    plt.show() 
```

```python
def subgrupos(data,nc=2,sub=30):
    new_data= []
    indices=[np.random.choice(data.shape[1],size=sub,replace=False) for i in range(nc)]
    for i in range(nc):
        new_data.append((data[i][indices[i]]))
    return new_data
```

```python
Arch2={llave:subgrupos(np.array(DatosG[llave])) for llave in neuronas}#diccionario de neuronas(datos sínteticos)
```

```python
for neuron in neuronas:
    A_real=ROC(DatosG[neuron][0],Arch2[neuron][0])
    A_permutada=permutacion(np.array(Arch2[neuron]))
    p=np.sum(A_permutada[A_permutada>A_real])/len(A_permutada)
    fig = plt.figure(figsize=(8,5))
    plt.hist(A_permutada,label='p='+str(p))
    plt.xlabel("bin",fontsize=14)
    plt.ylabel("frecuencia",fontsize=14)
    plt.title("Histograma AUROCs permutada "+'\n set '+neuron)
    plt.legend(fontsize="medium",shadow=True,fancybox=True)
    plt.show() 
```

```python
subgr=[n for n in range(10,200+10,10)]
```

```python
p_sub=[[]for neuron in neuronas]
ROCs_sub=[[]for neuron in neuronas]
for sub in subgr:
    Archn={llave:subgrupos(np.array(DatosG[llave]),nc=2,sub=sub) for llave in neuronas}#diccionario de neuronas(datos sínteticos)
    for i in range(len(neuronas)):
        A_real=ROC(np.array(Archn[neuronas[i]])[0],np.array(Archn[neuronas[i]])[1])
        A_permutada=permutacion(np.array(Archn[neuronas[i]]))
        p=np.sum(A_permutada[A_permutada>A_real])/len(A_permutada)
        p_sub[i].append(p)
        ROCs_sub[i].append(A_real)
```

```python
for i in range(len(neuronas)):
    fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,} # formato para el título
    fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,sharex=False,figsize=(15,5))
    ax1.scatter(subgr,p_sub[i])
    ax1.plot(subgr,p_sub[i])
    ax1.set_xticks(subgr)
    ax1.set_ylabel('p-value')
    ax1.set_xlabel('tamaño muestra')
    ax1.set_title('p-value vs tamaño de subgrupo\n set'+neuronas[i], fontdict_title) 
    ax2.scatter(subgr,ROCs_sub[i])
    ax2.plot(subgr,ROCs_sub[i])
    ax2.set_xticks(subgr)
    ax2.set_ylabel('ROC')
    ax2.set_xlabel('tamaño muestra')
    ax2.set_title('AUROC\n set'+neuronas[i], fontdict_title) 
```
