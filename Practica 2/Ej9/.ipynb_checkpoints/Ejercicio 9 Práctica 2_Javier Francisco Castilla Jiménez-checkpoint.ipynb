# Calculando la tasa de disparo 

```python
#Cargar los datos
NeuDPC1=nrn.importar('TiemposNeuDPC1.txt')
NeuDPC2=nrn.importar('TiemposNeuDPC2.txt')
NeuS1=nrn.importar('TiemposNeuS1.txt')
NeuS1A=nrn.importar('TiemposNeuS1A.txt')
#De estas primeras 4 Neuronas ya sé que cada clase/bloque tiene 15 ensayos
tiemposdet=[NeuDPC1,NeuDPC2,NeuS1,NeuS1A]
tasasdet=['NeuDPC1','NeuDPC2','NeuS1','NeuS1A']
```

```python
bS1_dt,nventanas=nrn.bordesup(0,8,0.01,0.05,nv=True)
def t2f(NeuS1):# esta función calcula los 
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(0,8,0.01,0.05,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=0.01,ventana=0.05,start=-2)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposdet)): # un ciclo por cada neurona 
    tasasdet[i]=t2f(tiemposdet[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasdet[0])# ahora los elementos de tasasS1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

```python
def nmk(k,r,t):
    suma=0
    for n in range(k+1):
        suma+=((r*t)**n)/math.factorial(n)
    p=(np.exp(-r*t))*suma
    return p
    
def nMk(k,r,t):
    nMk=1-nmk(k,r,t)
    return nMk
```

```python
probabilidad=[]
for tasa in range(len(tasasdet)): # ciclo para cada neurona
    espigas=(tasasdet[tasa]*0.05).astype(int) #convertimos la tasa en frecuencia de neuronas
    r=estimadobasal1[tasa]#tasa basal para esa neurona
    t=0.05# en este caso t que corresponde al tamaño de ventana es constante para todas las neuronas 
    prob=lambda espiga : nMk(espiga,r,t)# función anónima para calcular la probabilidad con los parametros especificos r t y k
##################################
    probas=[]# lista vacía con para las probabilidades de una neurona
    for i in range(len(espigas)): # este ciclo itera en el número de ensayos por neurona
        p=[prob(xi) for xi in espigas[i]] # realizamo las operación para obtener una probabilidad para cada elemento de cada ensayo de cada neurona 
        probas.append(p) # añadimos la lista de probabilidades para cada ventana de un ensayo a la lista probas
    probas=np.array(probas)
    probabilidad.append(probas)
###################################
```

```python
def responsividad(tasasdisparo,tbasal,ventana):
    probabilidad=[] #lista vacía para los arrays de probabilidades de cada neurona
    for tasa in range(len(tasasdisparo)): # ciclo para cada neurona
        espigas=(tasasdisparo[tasa]*ventana).astype(int) #convertimos la tasa en frecuencia de neuronas
        r=tbasal[tasa]#tasa basal para esa neurona
        t=ventana# en este caso t que corresponde al tamaño de ventana es constante para todas las neuronas 
        prob=lambda espiga : nMk(espiga,r,t)# función anónima para calcular la probabilidad con los parametros especificos r t y k
##################################
        probas=[]# lista vacía con para las probabilidades de una neurona
        for i in range(len(espigas)): # este ciclo itera en el número de ensayos por neurona
            p=[prob(xi) for xi in espigas[i]] # realizamo las operación para obtener una probabilidad para cada elemento de cada ensayo de cada neurona 
            probas.append(p) # añadimos la lista de probabilidades para cada ventana de un ensayo a la lista probas
        probas=np.array(probas)# array de las probabilidades de una neurona
        probabilidad.append(probas)# agregamos el array a la lista de arrays
    return probabilidad #lista de arrays donde cada array corresponde a las probabilidades de los ensayos de una neurona
###################################
```
