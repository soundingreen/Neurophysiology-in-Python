```python
import numpy as np 
import matplotlib.pyplot as plt
import NeuralFunctJC as nrn
```

```python
plt.style.use('seaborn')
```

```python
#carguemos los tiempos de las neuronas de S1
TS1st1=nrn.importar('TiemposNeuS1.txt')
TS1Ast1=nrn.importar('TiemposNeuS1A.txt')
TS1Bst1=nrn.importar2('TiemposNeuS1B.txt')
TS1Cst1=nrn.importar2('TiemposNeuS1C.txt')
TS1Dst1=nrn.importar2('TiemposNeuS1D.txt')
TS1Est1=nrn.importar2('TiemposNeuS1E.txt')
TS1Fst1=nrn.importar2('TiemposNeuS1F.txt')
TS1Gst1=nrn.importar2('TiemposNeuS1G.txt')
TS1Hst1=nrn.importar2('TiemposNeuS1H.txt')
TS1Ist1=nrn.importar2('TiemposNeuS1I.txt')
tiemposS1=[TS1st1,TS1Ast1,TS1Bst1,TS1Cst1,TS1Dst1,TS1Est1,TS1Fst1,TS1Gst1,TS1Hst1,TS1Ist1]# lista de listas con los np.ndarrays de lod tiempos de ocurrencia de las espigas
```

```python
tasasS1st1=['S1','S1A','S1B','S1C','S1D','S1E','S1F','S1G','S1H','S1I']# lista de nombres para las tasas de disparo
bS1_dt,nventanas=nrn.bordesup(0,1,1,1,nv=True)
print(len(bS1_dt))
def tasavNeu(NeuS1):
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(0,1,1,1,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=1,ventana=1,start=0)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposS1)): # un ciclo por cada neurona 
    tasasS1st1[i]=tasavNeu(tiemposS1[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasS1st1[1])# ahora los elementos de tasasS1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

```python
tasasS1st2=['S1','S1A','S1B','S1C','S1D','S1E','S1F','S1G','S1H','S1I']# lista de nombres para las tasas de disparoa
bS1_dt,nventanas=nrn.bordesup(3,4,1,1,nv=True)
print(len(bS1_dt))
def tasa_vNeu(NeuS1):
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(3,4,1,1,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=1,ventana=1,start=3)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposS1)): # un ciclo por cada neurona 
    tasasS1st2[i]=tasa_vNeu(tiemposS1[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasS1st2)# ahora los elementos de tasasS1st1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

```python
print(len(tasasS1st1))
print(len(tasasS1st2))
```

```python
#calculemos el z-score para el primer estímulo
zst1=[]
for i in range(len(tasasS1st1)):
    mu=np.mean(tasasS1st1[i])# tasa promedio para una neurona 
    sg=np.std(tasasS1st1[i])# desviación estándar para una neurona
    z=(tasasS1st1[i]-mu)/sg# zscore calculado usando mu y sigma de la tarea para una neurona
    zst1.append(z)
```

```python
#calculemos el z-score para el segundo estímulo
zst2=[]
for i in range(len(tasasS1st2)):
    mu=np.mean(tasasS1st2[i])# tasa promedio para una neurona 
    sg=np.std(tasasS1st2[i])# desviación estándar para una neurona
    z=(tasasS1st2[i]-mu)/sg# zscore calculado usando mu y sigma de la tarea para una neurona
    zst2.append(z)
```

```python
#función para separar de acuerdo a la indentidad del segundo estimulo
def separa1(Neu):
    sepagrup=lambda zscore :zscore[0:30]# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    sepaexten=lambda zscore :zscore[30:60]
    #tasa de disparo de cada ensayo de la neurona NeuS1
    Zagrupados=list(map(sepagrup,Neu)) # array con las tasas de disparo para cada ventana
    Zextendidos=list(map(sepaexten,Neu))
    return Zagrupados,Zextendidos
```

```python
zagrust1,zextenst1=separa1(zst1)
est1agrup=np.sort(np.concatenate(zagrust1))
est1exten=np.sort(np.concatenate(zextenst1))
```

```python
print(len(est1agrup))
print(len(est1exten))
```

```python
cuentas, bins=np.histogram(est1agrup,10)
punto_medio=(bins[0:-1]+bins[1:])/2
counts=(cuentas/len(est1agrup))/(bins[1]-bins[0])
plt.plot(punto_medio,counts,'bo:')
plt.hist(bins[:-1],bins,weights=counts)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Distrubución condicionada del z-score dado que el estímlo es agrupado \n Estímulo 1", fontdict_title)
plt.ylabel('probabilidad')
plt.xlabel('z-score')
plt.show()
```

```python
cuentas, bins=np.histogram(est1exten,10)
punto_medio=(bins[0:-1]+bins[1:])/2
counts=(cuentas/len(est1exten))/(bins[1]-bins[0])
plt.plot(punto_medio,counts,'bo:')
plt.hist(bins[:-1],bins,weights=counts)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Distrubución condicionada del z-score dado que el estímlo es extendido \n Estímulo 1", fontdict_title)
plt.ylabel('probabilidad')
plt.xlabel('z-score')
plt.show()
```

```python
#función para separar de acuerdo a la indentidad del segundo estimulo
def separa2(Neu):
    r=[i for i in range (15)]+[i for i in range(30,45)]
    r2=[i for i in range (15,30)]+[i for i in range(45,60)]
    sepagrup=lambda zscore :zscore[r]# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    sepaexten=lambda zscore :zscore[r2]
    #tasa de disparo de cada ensayo de la neurona NeuS1
    Zagrupados=list(map(sepagrup,Neu)) # array con las tasas de disparo para cada ventana
    Zextendidos=list(map(sepaexten,Neu))
    return Zagrupados,Zextendidos
```

```python
zagrust2,zextenst2=separa2(zst2)
est2agrup=np.sort(np.concatenate(zagrust2))
est2exten=np.sort(np.concatenate(zextenst2))
```

```python
cuentas, bins=np.histogram(est2agrup,10)
punto_medio=(bins[0:-1]+bins[1:])/2
counts=(cuentas/len(est2agrup))/(bins[1]-bins[0])
plt.plot(punto_medio,counts,'bo:')
plt.hist(bins[:-1],bins,weights=counts)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Distrubución condicionada del z-score dado que el estímlo es agrupado \n Estímulo 2", fontdict_title)
plt.ylabel('probabilidad')
plt.xlabel('z-score')
plt.show()
```

```python
cuentas, bins=np.histogram(est2exten,10)
punto_medio=(bins[0:-1]+bins[1:])/2
counts=(cuentas/len(est2exten))/(bins[1]-bins[0])
plt.plot(punto_medio,counts,'bo:')
plt.hist(bins[:-1],bins,weights=counts)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Distrubución condicionada del z-score dado que el estímlo es extendido \n Estímulo 2", fontdict_title)
plt.ylabel('probabilidad')
plt.xlabel('z-score')
plt.show()
```

## la distribución condicionada del z-score dado que el estímulo es agrupado para el estímulo 1 tiene su mayor probabilidad está alrededor de z=0 creo que eso se debe a que para calcular z-score solo use mu y sigma de esa de ventana de 1 segundo en el estímulo , sin embargo en la el caso del estímulo 2 también solo use esa ventana para calcular sigma y mu pero su mayor probabilidad es en un z-score ligeramente más alejado de cero, sin embargo la forma general de la distrubución de probabilidad para el estímulo 1 y para el estímulo dos dado que es estímulo es agrupado se parecen bastante entre si. Y eso es es algo que esperaba observar pues al final son las mismas neuronas y el mismo estímulo solo cambia el momento en que se le aplica el estímlo debido a esto creo que las pequeñas diferencias entre las dos distribuciones dado que el estímulo es agrupado se deben a la variabilidad neuronal intrinseca y no al momento en que se dió el estímulo.

## la distribución condicionada del z-score dado que el estímulo es extendido para el estímulo 1 es muy muy similar  al a la distribución para el estímulo dos dado que el estímulo es extendido. Y eso es es consistente con la idea de que lo que determina la distribución de probabilidad es el tipo de estímulo y no el momento en que se dió este estímulo (ya sea que se de como el estímulo 1 o como el estímulo 2), debido a que son las mismas neuronas recibiendo el mismo estímulo y lo único que cambia es el momento en que se le aplica el estímlo debido.
