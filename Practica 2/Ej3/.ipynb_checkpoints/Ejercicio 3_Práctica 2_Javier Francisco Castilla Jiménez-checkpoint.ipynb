```python
import numpy as np 
import matplotlib.pyplot as plt
import NeuralFunctJC as nrn
plt.style.use('seaborn')
```

```python
TS1A=nrn.importar2('Neu_Det_S1A.csv')
TS1B=nrn.importar2('Neu_Det_S1B.csv')
TS1C=nrn.importar2('Neu_Det_S1C.csv')
TS1D=nrn.importar2('Neu_Det_S1D.csv')
TS1E=nrn.importar2('Neu_Det_S1E.csv')
tiemposS1=[TS1A,TS1B,TS1C,TS1D,TS1E]# lista de listas con los np.ndarrays de los tiempos de ocurrencia de las espigas
```

```python

bqA=list((nrn.bloques(nrn.abrir('Neu_Det_S1A.csv'))))
bqB=list((nrn.bloques(nrn.abrir('Neu_Det_S1B.csv'))))
bqC=list((nrn.bloques(nrn.abrir('Neu_Det_S1C.csv'))))
bqD=list((nrn.bloques(nrn.abrir('Neu_Det_S1D.csv'))))
bqE=list((nrn.bloques(nrn.abrir('Neu_Det_S1E.csv'))))

bqs={'A':bqA,'B':bqB,'C':bqC,'D':bqD,'E':bqE}
bqs
```

```python
tasasS1=['S1A','S1B','S1C','S1D','S1E']# lista de nombres para las tasas de disparo
bS1,nventanas=nrn.bordesup(-2,3.5,0.01,0.05,nv=True)
def t2frS1(NeuS1):# esta función calcula los 
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(-2,3.5,0.01,0.05,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=0.01,ventana=0.05,start=-2)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposS1)): # un ciclo por cada neurona 
    tasasS1[i]=t2frS1(tiemposS1[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasS1[1])# ahora los elementos de tasasS1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

# pues la actividad es bastante homognea sin embargo parece que la neurona D presenta un mayor variabilidad en términos generales.

```python
kns1='ABCDE'#clave de letras para las neuronas
muS1Tarea={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
sgS1Tarea={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
zscoresS1={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
bl1,bl2,bl3,bl4,bl5,bl6=0,0,0,0,0,0
sg1,sg2,sg3,sg4,sg5,sg6=0,0,0,0,0,0
r1,r2,r3,r4,r5,r6=0,0,0,0,0,0
j1,j2,j3,j4,j5,j6=0,0,0,0,0,0
zbloquesS1=[bl1,bl2,bl3,bl4,bl5,bl6]
zblqsigma=[sg1,sg2,sg3,sg4,sg5,sg6]
rj=[r1,r2,r3,r4,r5,r6]
sj=[j1,j2,j3,j4,j5,j6]

mpobla,spobla,zprom=0,0,0

for i in range(len(tasasS1)):
    mu=np.mean(tasasS1[i])# tasa promedio en toda la tarea para una neurona 
    muS1Tarea[kns1[i]]=mu
    sg=np.std(tasasS1[i])# desviación estándar en toda la tarea para una neurona
    sgS1Tarea[kns1[i]]=sg
    z=(tasasS1[i]-mu)/sg# zscore calculado usando mu y sigma de la tarea para una neurona
    zscoresS1[kns1[i]]=z
    mpobla+= mu
    spobla+=sg
    cl1,cl2,cl3,cl4,cl5,cl6=nrn.clases(z,bqs[kns1[i]])
    clase=[cl1,cl2,cl3,cl4,cl5,cl6]
    for j in range(len(clase)):
        zbloquesS1[j]+=np.mean(clase[j],0)/len(tasasS1)
        zblqsigma[j]+=np.std(clase[j],0)/len(tasasS1)
        
spobla=spobla/len(tasasS1)
mpobla=mpobla/len(tasasS1)
#zprom=zprom/len(tasasS1)

##tasa poblacional
for j in range(len(zbloquesS1)):
    rj[j]=(zbloquesS1[j]*spobla)+mpobla
```

```python
ej_a='BD'
for key in ej_a:
    ztmc,ztsgc=nrn.clases_Z(zscoresS1[key],bqs[key])
    titulo='z-score usando mu y sigma de la tarea\n por clase neurona S1{0} \n tarea de discriminación táctil'.format(key)
    nrn.graf6c(bS1,ztmc,ztsgc,titulo)
```

```python
ej_c='ABCDF'
for key in ej_c:
    titulo='z-score promedio\n por clase neurona S1{0} \n tarea de discriminación táctil'.format(key)
    nrn.graf6c(bS1,zbloquesS1,zblqsigma,titulo)
```

```python
titulo='tasa promedio poblacional usando mu y sigma basales\n por clase neurona S1{0} \n tarea de discriminación táctil'.format(key)
nrn.sixgr(bS1,rj,titulo,'tasa[Hz]')
```
