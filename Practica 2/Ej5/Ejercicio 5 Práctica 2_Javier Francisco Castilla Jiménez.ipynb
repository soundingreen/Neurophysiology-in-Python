# realmente con 10000 números de tamaño de muestra no se nota con tanta claridad la convergencia y sin embargo se alcanza a notar que la probabilidad converge a un número que está entre 3.6 y 3.8 , quizá más cerca de 3.7

la idea es que debido a que todos los número entre 1 y 300 tienen la misma probabilidad de salir es decir son equiprobables, el número al que converge debe ser 1 menos la probabilidad de que un número sea igual o menor que 190 es decir si tomamos una muestra discreta de enteros de 1 a 300 podemos calcular (aproximar) la probabilidad a la que converge como $$p=1-\left(\frac{1}{300}\cdot{190}\right)=0.3666666666$$ 

```python
probabilidades=[]
tamaños=np.arange(100,10000+100,100)
for tamaño in tamaños:
    Aleatorios=(np.random.rand(tamaño))*300
    p=np.sum(Aleatorios>190)/tamaño
    probabilidades.append(p)

probabilidades=np.array(probabilidades)
```

```python
plt.scatter(tamaños,probabilidades)
plt.xlabel('tamaño de muestra')
plt.ylabel('probabilidad')
plt.title('Probabilidad de tener un número mayor que 190\n en función del tamaño de muestra')
```
