```python
bS1_dtv1,nventanas=nrn.bordesup(-0.5,1.5,0.01,0.05,nv=True)
print(nventanas)
def t2frS1(NeuS1):# esta función calcula los 
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(-0.5,1.5,0.01,0.05,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=0.01,ventana=0.05,start=-0.5)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposS1)): # un ciclo por cada neurona 
    tasasS1st1v1[i]=t2frS1(tiemposS1[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasS1st1v1[1])# ahora los elementos de tasasS1st1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

# Agrupado de acuerdo a la identidad del primer estímulo

```python
#función para separar de acuerdo a la indentidad del segundo estimulo
def st1(Neu):
    sepagrup=lambda tasa :tasa[0:30]# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    sepaexten=lambda tasa:tasa[30:60]
    #tasa de disparo de cada ensayo de la neurona NeuS1
    agrupados=list(map(sepagrup,Neu)) # array con las tasas de disparo para cada ventana
    extendidos=list(map(sepaexten,Neu))
    return agrupados,extendidos
```

# Ventanas de 50 ms y pasos de 10 ms

```python
st1grouped,st1extended=st1(tasasS1st1v1)
```

```python
def fano(mu,var):
    """
    Dado un array de tasas, una array de medias (o un array renglón) y 
    un array de std (o un array renglón).Esta función calcula el z-score,
    y devuelve una matriz de z-score's.
    """
    if np.sum(mu==0)==0:
        f=var/mu
    else:
        mu[mu==0]=1
        f=var/mu
    return f
```

```python
fanost1grouped=[]
for i in range(len(st1grouped)):
    mu=np.mean(st1grouped[i],0)# tasa promedio para una neurona 
    var=np.var(st1grouped[i],0)# desviación estándar para una neurona
    f=fano(mu,var) # factor de fano
    fanost1grouped.append(f)
fpromst1grouped=np.mean(np.stack(fanost1grouped),0)
```

```python
plt.plot(bS1_dtv1,fpromst1grouped)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Factor de Fano promedio para estímulo agrupado \n Estímulo 1 \n 10 neuronas de S1\n paso=0.01 y ventana=0.05", fontdict_title)
plt.ylabel('factor de fano')
plt.xlabel('tiempo[s]')
plt.show()
```

```python
fanost1extended=[]
for i in range(len(st1extended)):
    mu=np.mean(st1extended[i],0)# tasa promedio para una neurona 
    var=np.var(st1extended[i],0)# desviación estándar para una neurona
    std=np.std(st1extended[i],0)
    f=fano(mu,var) # factor de fano
    fanost1extended.append(f)
fanost1extended=np.mean(np.stack(fanost1extended),0)
```

```python
plt.plot(bS1_dtv1,fanost1extended)
fontdict_title = {'family': 'serif','color':  'darkred','weight': 'normal','size': 16,}
plt.title("Factor de Fano promedio para estímulo extendido \n Estímulo 1 \n 10 neuronas de S1\n paso=0.01 y ventana=0.05", fontdict_title)
plt.ylabel('factor de fano')
plt.xlabel('tiempo[s]')
plt.show()
```

En contraste con esto con ventanas más grandes como lo  son ventanas de 200 ms y pasos de 40 ms, como la diferencia de ventana a ventana baja entonces es menos variable, la varianza baja y el factor de Fano también lo hace.
