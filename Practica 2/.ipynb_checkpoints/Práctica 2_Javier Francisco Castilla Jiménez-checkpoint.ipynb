```python
content=nrn.abrir('Neu_Det_S1B.csv')
contador=0 
clases=[] #lista vacía para guardar variables
for string in content:# itero los elementos
    s=string.strip().split(",") # al hacer esto identifico los separadores aunque no sean \n
    if s.count('')==len(s): # si es un separador de bloques
        clases.append(contador) #agrego acumulado en contandor a la lista clases
        contador=0# reinicio el contador
    else:# en la mayoría de los casos donde no es un separador
        contador+=1 #sumo 1 al contador

clases.append(contador)# agrego el acumulado despues del último contador
bloques=np.array(clases)# lo convierto en un array por si necesito operar con los datos
print(bloques)

# esto lo generalize en forma de una función y lo agregue a mi librería personal de funciones de procesamiento
```
