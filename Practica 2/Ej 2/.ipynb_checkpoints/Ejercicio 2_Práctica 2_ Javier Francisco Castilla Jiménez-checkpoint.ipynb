```python
import numpy as np 
import matplotlib.pyplot as plt
import NeuralFunctJC as nrn
```

```python
plt.style.use('seaborn')
```

```python
#carguemos los tiempos de las neuronas de S1
TS1=nrn.importar('TiemposNeuS1.txt')
TS1A=nrn.importar('TiemposNeuS1A.txt')
TS1B=nrn.importar2('TiemposNeuS1B.txt')
TS1C=nrn.importar2('TiemposNeuS1C.txt')
TS1D=nrn.importar2('TiemposNeuS1D.txt')
TS1E=nrn.importar2('TiemposNeuS1E.txt')
TS1F=nrn.importar2('TiemposNeuS1F.txt')
TS1G=nrn.importar2('TiemposNeuS1G.txt')
TS1H=nrn.importar2('TiemposNeuS1H.txt')
TS1I=nrn.importar2('TiemposNeuS1I.txt')
tiemposS1=[TS1,TS1A,TS1B,TS1C,TS1D,TS1E,TS1F,TS1G,TS1H,TS1I]# lista de listas con los np.ndarrays de lod tiempos de ocurrencia de las espigas
```

```python
tasasS1=['S1','S1A','S1B','S1C','S1D','S1E','S1F','S1G','S1H','S1I']# lista de nombres para las tasas de disparo
bS1_dt,nventanas=nrn.bordesup(-2,8,0.01,0.05,nv=True)
def t2frS1(NeuS1):# esta función calcula los 
    ## bordes para neurona de S1
    bS1_dt,nventanas=nrn.bordesup(-2,8,0.01,0.05,nv=True) #array con los bordes superiores de cada ventana
    #calculo de tasa de disparo ensayo por ensayo
    func=lambda ensayo : nrn.tasa(ensayo,nventanas=nventanas,paso=0.01,ventana=0.05,start=-2)# esta función calcula la tasa de disparo de un ensayo nventanas, paso=0.01,ventana=0.05,start=-2 
    #tasa de disparo de cada ensayo de la neurona NeuS1
    tasasNeuS1=map(func,NeuS1) # array con las tasas de disparo para cada ventana
    tasasNeuS1=np.array(list(tasasNeuS1))# este array contiene las tasas de disparo de cada ensayo de esta neurona
    return  tasasNeuS1 # devuelve las tasas de disparo en un array

for i in range(len(tiemposS1)): # un ciclo por cada neurona 
    tasasS1[i]=t2frS1(tiemposS1[i]) # a partir de los tiempos calcula las tasas de disparo

type(tasasS1[1])# ahora los elementos de tasasS1 son np.ndarrays con las tasas de disparo de una neurona. # tasasS1 ahora es una lista de ndarrays
```

# observo que las neuronas A,B,C parecen tener mucho más ruido en su actividad, mientras están en reposo mientras que D y F parecen ser completamente silentes antes y después de la llegada del estímulo.

```python
kns1='1ABCDEFGHI'#clave de letras para las neuronas
muS1Tarea={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
sgS1Tarea={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
zscoresS1={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
mpobla,spobla,zprom=0,0,0

for i in range(len(tasasS1)):
    mu=np.mean(tasasS1[i])# tasa promedio en toda la tarea para una neurona 
    muS1Tarea[kns1[i]]=mu
    sg=np.std(tasasS1[i])# desviación estándar en toda la tarea para una neurona
    sgS1Tarea[kns1[i]]=sg
    z=(tasasS1[i]-mu)/sg# zscore calculado usando mu y sigma de la tarea para una neurona
    zscoresS1[kns1[i]]=z
    mpobla+= mu
    spobla+=sg
    zprom+=z
    

spobla=spobla/len(tasasS1)
mpobla=mpobla/len(tasasS1)
zprom=zprom/len(tasasS1)
```

```python
ej_a='ABCDF'
for key in ej_a:
    ztmc,ztsgc=nrn.clases_Z(zscoresS1[key],(15,15,15,15))
    titulo='z-score usando mu y sigma de la tarea\n por clase neurona S1{0} \n tarea de discriminación táctil'.format(key)
    nrn.graf4c(bS1_dt,ztmc,ztsgc,titulo)
    
```

```python
z_prom_pob_clase,zsgp=nrn.clases_Z(zprom,(15,15,15,15))
nrn.graf4c(bS1_dt,z_prom_pob_clase,zsgp,'z-score promedio (de 10 neuronas de S1)\n por clase \n tarea de discriminación táctil')
```

```python
z_prom_pob_clase,zsgp=nrn.clases_Z((zprom*spobla)+mpobla,(15,15,15,15))
nrn.c4plot(bS1_dt,z_prom_pob_clase,zsgp,'tasa promedio promedio poblacional \n(de 10 neuronas de S1)\n por clase \n tarea de discriminación táctil',ejey='tasa[Hz]')
```

# Usar la media y la sigma basales para cálcular el z-score resalta la actividad mientras más se aleja de la actividad basal esto nos permite tener más claridad y al observar la actividad frente a  la  llegada del estímulo, pues mientras al usar mu y sigma basales estás neuronas alcanzan una actividad de 10,20 y hasta 40 z-scores respectivamente, mientras que las neuronas del inciso pasado donde se usaban mu y sigma de toda la tarea no iban más allá de los 10 z-score

```python
bsa,nvbasalesS1=nrn.bordesup(-2,0,0.01,0.05,nv=True) #array con los bordes superiores de cada ventana
kns1='1ABCDEFGHI'#clave de letras para las neuronas
muS1basal={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
sgS1basal={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
zscoresS1bas={a:0 for a in kns1}# diccionario con claves de letras para las neuronas
mpoblab,spoblab,zpromb=0,0,0
for i in range(len(tasasS1)):
    mub=np.mean(tasasS1[i][:,:nvbasalesS1])# tasa promedio en toda la tarea para una neurona 
    muS1basal[kns1[i]]=mub
    sgb=np.std(tasasS1[i][:,:nvbasalesS1])# desviación estándar en toda la tarea para una neurona
    sgS1basal[kns1[i]]=sgb
    zb=(tasasS1[i]-mub)/sgb# array zscore calculado usando mu y sigma de la tarea para una neurona
    zscoresS1bas[kns1[i]]=zb
    mpoblab+= mub
    spoblab+=sgb
    zpromb+=zb
    

spoblab=spoblab/len(tasasS1)
mpoblab=mpoblab/len(tasasS1)
zpromb=zpromb/len(tasasS1)
```

```python
ej_c='ABCDF'
for key in ej_c:
    zbmc,zbsgc=nrn.clases_Z(zscoresS1bas[key],(15,15,15,15))
    titulo='z-score usando mu y sigma basales\n por clase neurona S1{0} \n tarea de discriminación táctil'.format(key)
    nrn.graf4c(bS1_dt,zbmc,zbsgc,titulo)
    
```

```python
z_prom_pob_claseb,zsgpb=nrn.clases_Z(zpromb,(15,15,15,15))
nrn.graf4c(bS1_dt,z_prom_pob_claseb,zsgpb,'z-score promedio (de 10 neuronas de S1)\n usando mu sigma basales\n por clase \n tarea de discriminación táctil')
```

```python
z_prom_pob_claseb,zsgpb=nrn.clases_Z((zpromb*spoblab)+mpoblab,(15,15,15,15))
nrn.c4plot(bS1_dt,z_prom_pob_claseb,zsgpb,'tasa promedio promedio poblacional \n(de 10 neuronas de S1)\n por clase \n tarea de discriminación táctil',ejey='tasa[Hz]')
```
