```python
Dat_Ej_A=[Data50g0,Data200g0,Data500g0,Data200g2]
Dat_Ej_B=[Data200P2,Data200P4]
def RMS(array):# función para calcular el valor medio cuadrático
    rms=np.sqrt(np.sum(np.power(array,2)/len(array)))
    return rms
def tercerm(array):# función para calcular el tercer momento centrado
    delta=array-np.mean(array)
    tercer=(np.power(delta,3))
    tercer=np.sum(tercer)/len(array)
    return tercer 
def cuartom(array):# función para calcular el cuarto momento centrado
    delta=array-(np.mean(array))
    cuarto=(delta**4)
    cuarto1=np.sum(cuarto)/len(array)
    return cuarto1


def ejercicio4(array):
    mu=np.mean(array)
    var=np.var(array)
    std=np.std(array)
    rms=RMS(array)
    mo3=tercerm(array)
    sk=skew(array)
    mo4=cuartom(array)
    kur_f=kurtosis(array,fisher=True)
    return mu,var,std,rms,mo3,sk,mo4,kur_f

```

```python
def gauss(x,mu,sigma): #x son mis datos en x(los bines),a es un factor de ajuste porque los datos no están normalizados y sino el area debajo de la gaussiana siempre vale uno,mu es la media y sigma la std 
    return (1/(sigma*np.sqrt(2*math.pi)))*np.exp(-(x-mu)**2/(2*sigma**2))# gaussian para ajuste
```

```python
nombres=['Data50g0','Data200g0','Data500g0','Data200g2']
Bins=[11,15,20,20]
for i in range(len(Dat_Ej_A)):
    mu=estadisticos[i][0]#media
    sg=estadisticos[i][2]#sigma
    cuentas, bins=np.histogram(Dat_Ej_A[i],bins=Bins[i]) #count guarda un array con la frecuencia y bins un array con los bordes de bins
    w=bins[2]-bins[1]# el ancho de bin
    suma=cuentas.sum() # número total de datos
    counts=(cuentas/suma)/w #normalizamos los datos dividiendo la frecuencia entre el número total de datos y el ancho de la ventana
    plt.hist(bins[:-1],bins,weights=counts)# siguiendo la documentación creamos un histograma usando    plt.hist a partir de datos en obtenidos vía np.histogram
    x=bins[1:]
    y=counts
    
    popt,pcov = curve_fit(gauss,x,y,p0=[mu,sg]) # calculamos los valores ajustados
    plt.plot(x,counts,'b+:',label='datos')
    plt.plot(x,gauss(x,*popt),'ro:',label='ajuste')
    plt.title(nombres[i])
    plt.legend()
    plt.show()
    
    ##########
    w=bins[2]-bins[1]# el ancho de bin
    media,sigma=popt
    tercer_mom=(media**3)+(3*media*(sigma**2))
    cuarto_momc=3*(sigma**4)
    ####
    print('Estadisticos de {}:'.format(nombres[i]))
    print('número total de datos:',suma)
    print('mu:',media,'\n sigma:',sigma)
    print('tercer_momento:',tercer_mom,'tercer momento central',0)
    print('cuarto_momento_central:',cuarto_momc)
    print('bins:',Bins[i])
    print('ancho de ventana',w)#imprimos el ancho de bin
    print('#####################################################')
```

```python
#función para obtener los datos
def frag(fragmento):
    with open(fragmento,mode='r') as libro1: # observemos que el txt tiene que estar en la misma carperta que el script si voy usar esta sintaxis
        contenido1= libro1.readlines() #contenido1 es una list cuyo unico elemento es un string con todo el texto del libro
    texto=contenido1[0]#aquí ya cerré el archivo txt y tengo su contenido almacenado de un  string que llame cosmos 
    for c in ".,()[]-:;¿?": # con este ciclo reemplazo todos los signos de puntuación que aparecen en los textos
        texto= texto.replace(c,"")
    textosa=texto.split() #con el metodo split creo listas que contienen las palabras en forma de str separadas como elementos de una lista
    valores, frecuencia=np.unique(textosa, return_counts=True) # contamos los valores unicos y los asignasmos a lista de valores # contamos la frecuencia de cada valor unico y lo asignamos a 
    total=np.sum(frecuencia)
    orden=np.argsort(frecuencia)
    frecuencia=frecuencia[orden]
    relativa=frecuencia/total #divimos la frecuencia entre el número total de palabras
    return valores,frecuencia,relativa
```

```python
# en forma de función para automatizar los siguientes cálculos.
def est_words(palabras,probabilidad):
    """los arreglos deben estar ordenados para que funcione está función"""
    position=np.array([i for i in range (1,len(palabras)+1)])
    media=np.sum(position*probabilidad)
    mensaje1=print('la media de las palabras en mi fragmentoes:',media)
    var=np.sum(np.power((position-media),2)*probabilidad)
    mensaje2=print('la varianza de las palabras en mi fragmento es: {:.2f}'.format(var))
    std=np.sqrt(var1)
    mensaje3=print('la desviación estándar de las palabras en mi fragmento son: {:.2f}'.format(std))
    return media,var,std
```
