```python
Data50g0=np.loadtxt('Data50g0.txt')
Data200g0=np.loadtxt('Data200g0.txt')
Data500g0=np.loadtxt('Data500g0.txt')
Data200g2=np.loadtxt('Data200g2.txt')
Data200P2=np.loadtxt('Data200P2.txt')
Data200P4=np.loadtxt('Data200P4.txt')
Datos=[Data50g0,Data200g0,Data500g0,Data200g2,Data200P2,Data200P4]

```

```python
for lista in Datos: # comprobamos que la importación de datos fue correcta y que no se perdieron datos.
    print(len(lista))
```

```python
Dat_Ej_A=[Data50g0,Data200g0,Data500g0,Data200g2]
def RMS(array):# función para calcular el valor medio cuadrático
    rms=np.sqrt(np.sum(np.power(array,2)/len(array)))
    return rms
def tercerm(array):# función para calcular el tercer momento centrado
    delta=array-np.mean(array)
    tercer=(np.power(delta,3))/len(array)
    tercer=np.sum(tercer)/len(array)
    return tercer 
def cuartom(array):# función para calcular el cuarto momento centrado
    delta=array-np.mean(array)
    cuarto=(np.power(delta,4))
    cuarto=np.sum(cuarto)/len(array)
    return cuarto 

def ejercicio4(array):
    mu=np.mean(array)
    var=np.var(array)
    std=np.std(array,ddof=1)
    rms=RMS(array)
    mo3=tercerm(array)
    mo4=cuartom(array)
    return mu,var,std,rms,mo3,mo4
```

```python
estadisticos=[]
for lista in Dat_Ej_A:
    estadisticos.append(ejercicio4(lista))
```

```python
nombres=['Data50g0','Data200g0','Data500g0','Data200g2']
for i in range(len(Dat_Ej_A)):
    mu,var,std,rms,mo3,mo4=estadisticos[i]
    print(nombres[i],'\n media:',mu,'\n varianza:',var,'\n std:',std,'\n valor medio cuadrático:',rms,'\n 3er_momento centrado:',mo3,'\n 4to_momento centrado:',mo4)
```

```python
mean=estadisticos[0][0]
sg=estadisticos[0][2]
counts, bins=np.histogram(Data50g0,bins=10) #count guarda un array con la frecuencia y bins un array con los bordes de bins
plt.hist(bins[:-1],bins,weights=counts)# siguiendo la documentación creamos un histograma usando plt.hist a partir de datos en obtenidos vía np.histogram
x=(bins[0:-1]+bins[1:])/2
y=counts

def gauss(x,a,x0,sigma):
       return a*exp(-(x-x0)**2/(2*sigma**2))

popt,pcov = curve_fit(gauss,x,y,p0=[1,mean,sg])

plt.plot(x,counts,'b+:',label='data')
plt.plot(x,gauss(x,*popt),'ro:',label='fit')
plt.legend()

w=bins[2]-bins[1]# el ancho de bin
print('ancho de ventana',w)#imprimos el ancho de bin
```
