```python
def Distributions_bvst(zbasal,ztot,nv,nbins=200):
    mi=np.floor(min((zbasal.min(),ztot.min())))# el mínimo de mínimos de las dos distribuciones
    mx=np.ceil(max((zbasal.max(),ztot.max())))# el máximo de máximos de las dos distribuciones
    bins=np.linspace(mi,mx,nbins) # genero un linspace entre el minimo de minimos y el maximo de maximos, y este linspace va a tener el número de bines que yo indique en nbins 
    nv=len(ztot[0])# cuantos datos tengo en el sample y que me va a servir para normalizar las frecuencias y convertirlas en probabilidades
    m,n=zbasal.shape
    dist1=np.histogram(zbasal,bins=bins)[0]/(m*n) #probabilidades de los bines de la muestra 1
    dist2=[]
    for v in range(nv):
        dist=np.histogram(ztot[:,v],bins=bins)[0]/(len(ztot[:,v])) #probabilidades de los bines de la muestra 2
        dist2.append(dist)
    dist2=np.vstack(dist2)  
    return dist1,dist2,bins
```

```python
def latrep(roc,tiempos):
    comp=roc>=0.6
    for i in range(len(roc)):
        if comp[i]==True and comp[i]==comp[i+1]==comp[i+2]:
            break
    return tiempos[i]
```

```python
# todas las tasas de disparo en diccionario de neuronas
Ts=nrn.dicneuronrate(tiemposS1,neuronas,start=-2,stop=3.5,paso=0.01,ventana=0.05)
ms=dict() #diccionario para la media y la sigma totales de una neurona
for neuron in neuronas:
    mu=np.mean(Ts[neuron])
    sg=np.std(Ts[neuron])
    ms.update({neuron:(mu,sg)}) 
############## tasas de disparo en periodos 0 a 2s y 3 a 5s
Ts_p1=nrn.dicneuronrate(tiemposS1,neuronas,start=-1,stop=0,paso=0.01,ventana=0.05)
Ts_p2=nrn.dicneuronrate(tiemposS1,neuronas,start=0,stop=3,paso=0.01,ventana=0.05)
```

```python
distribuciones=dict()
for neuron in neuronas:
    basal=Ts_p1[neuron]
    tasal=Ts_p2[neuron]
    distbas,distasa,bines=Distributions_bvst(basal,tasal,nv=9600,nbins=100)
    distribuciones.update({neuron:{'basal':distbas,'est':distasa,'bines':bines}})
```

```python
ROCs=dict()
for neuron in neuronas:
    r=[]
    for v in range(len(Ts_p2['A'][0])):
        dist1=distribuciones[neuron]['basal']
        dist2=distribuciones[neuron]['est']
        auroc=nrn.getAUROC(dist1,dist2[v])
        r.append(auroc)
    ROCs.update({neuron:r})
    
```
